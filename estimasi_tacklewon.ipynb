```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('defending.csv')
```

# Visualisasi Data

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=True)
```

# Jumlah bermain berdasarkan posisi

```python
models = df.groupby('position').count()[['match_played']].sort_values(by='match_played',ascending=True).reset_index()
models = models.rename(columns={'match_played':'numberofposition'})
```

```python
fig = plt.figure(figsize=(15,5))
sns.barplot(x=models['position'], y=models['numberofposition'], color='royalblue')
plt.xticks(rotation=60)
```

# distribusi jumlah tackle

```python
plt.figure(figsize=(15,5))
sns.distplot(df['tackles'])
```

# SELEKSI FITUR

```python
features = ['balls_recoverd', 't_won', 't_lost', 'clearance_attempted', 'match_played']
x = df[features]
y = df['tackles']
x.shape, y.shape
```

# split data training dan data testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x,y,random_state=70)
y_test.shape
```

# membuat model regresi linier


```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
pred = lr.predict(X_test)
```

```python
score = lr.score(X_test, y_test)
print('akurasi model regresi linier = ', score)
```

# membuat inputan model regresi linier


```python
#balls_recoverd'=76, 't_won'=13, 't_lost'18, 'clearance_attempted'19, 'match_played=11
input_data = np.array([[76,	13,	18,	19,	11]])

prediction = lr.predict(input_data)
print('Estimasi tackle yang dimenangi oleh posisi defender', prediction)
```
